```python
import numpy as np
import matplotlib.pyplot as plt
import random

# b in 1e-2~2e-1
b = random.uniform(1e-2, 2e-1)
# k in 1e-3~1e-2
k = random.uniform(1e-3, 1e-2)
print(f"b={b}, k={k}")

delta_t = 0.01
t_max = 1000

# s = np.arange(0, t_max, delta_t)
# i = np.arange(0, t_max, delta_t)
# r = np.arange(0, t_max, delta_t)

s = np.zeros(int(t_max/delta_t))
i = np.zeros(int(t_max/delta_t))
r = np.zeros(int(t_max/delta_t))

s[0] = 1
i[0] = 1e-6
r[0] = 0
```

```python
for t in range(1, len(s)):
    s[t] = s[t-1] - b * s[t-1] * i[t-1] * delta_t
    i[t] = i[t-1] + (b * s[t-1] * i[t-1] - k * i[t-1]) * delta_t
    r[t] = r[t-1] + k * i[t-1] * delta_t
```

```python
plt.plot(s, label='Susceptible')
plt.plot(i, label='Infected')
plt.plot(r, label='Recovered')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('SIR Model Simulation')
plt.legend()
plt.grid()
plt.savefig('sir_model_simulation.png', dpi=400)
plt.show()
```
