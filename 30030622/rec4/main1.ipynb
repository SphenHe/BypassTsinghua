```python
import numpy as np

# 2D Possion Equation

n = 50

f = np.ones(n**2)

A = np.zeros((n**2, n**2))

for i in range(n**2):
    A[i][i] = 4
    if i + 1 < n**2:
        A[i][i+1] = -1
    if i - 1 >= 0:
        A[i][i-1] = -1
    if i + n < n**2:
        A[i][i+n] = -1
    if i - n >= 0:
        A[i][i-n] = -1

for i in range(n):
    A[i*n][i*n] = 2

A = A*n**2

print(A)
```

```python
u = np.linalg.solve(A, f)
```

```python
import matplotlib.pyplot as plt
u = u.reshape((n, n))
plt.imshow(u, extent=[0,1,0,1], cmap='jet', interpolation='nearest')
plt.title('2D Poisson Equation')
plt.colorbar()
plt.savefig('2D_Poisson.png', dpi=500)
plt.show()

print(u)
```
